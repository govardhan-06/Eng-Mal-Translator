### Preparing custom dataset

```python
from datasets import get_dataset_config_names, load_dataset

#checking the supported languages by the model
get_dataset_config_names("facebook/flores")
```

```python
dataset = load_dataset("facebook/flores", "eng_Latn-mal_Mlym")
dataset
```

```python
from datasets import Dataset
custom_dataset={'sentence_eng_Latn': dataset['devtest']['sentence_eng_Latn']+dataset['dev']['sentence_eng_Latn'],
           'sentence_mal_Mlym': dataset['devtest']['sentence_mal_Mlym']+dataset['dev']['sentence_mal_Mlym']}

custom_dataset=Dataset.from_dict(custom_dataset)
custom_dataset
```

```python
split_dataset=custom_dataset.train_test_split(test_size=0.1,seed=42)
split_dataset
```
